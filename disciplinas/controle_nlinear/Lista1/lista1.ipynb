```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
import matplotlib.gridspec as gridspec

h = 1.5
w = 1.5
delta = 0.1

x = np.arange(-w, w, delta)
y = np.arange(-h, h, delta)
X, Y = np.meshgrid(x, y)

# Dynamic system
U = Y
V = -2*X - 2*Y - 4*X**3
speed = np.sqrt(U**2 + V**2)

# Ljapunov function
L = 4*X**2 + 2*Y**2 + 4*X**4;

plt.rc('text', usetex=True)
plt.rc('font', family='serif')

fig = plt.figure(num=None, figsize=(30, 30), dpi=80, 
                 facecolor='w', edgecolor='k')
gs = gridspec.GridSpec(nrows=3, ncols=2, height_ratios=[1, 1, 2])

ax0 = fig.add_subplot(gs[0, 0])
ax0.streamplot(X, Y, U, V, density=4, color='k')
ax0.set_title(r'Plano de fase - $\dot x_1 = x_2$, $\dot x_2 = -2x_1 -2x_2-4x_1^3$')

ax0.contour(X, Y, L, levels = [0.5, 1.25, 2.5, 5, 10, 15, 20, 25], 
                     colors='blue', 
                     linestyles='dashed')

ax0.set_xlabel('$x_1$');
ax0.set_ylabel('$x_2$');

ax = plt.gca()
ax.set_xlim(-w, w)
ax.set_ylim(-h, h)

plt.gca().set_aspect('equal', adjustable='box')

plt.savefig('ex1.eps', bbox_inches="tight", pad_inches=0)
plt.savefig('ex1.png', bbox_inches="tight", pad_inches=0)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
import matplotlib.gridspec as gridspec

h = 4
w = 4
delta = 0.1

x = np.arange(-w, w, delta)
y = np.arange(-h, h, delta)
X, Y = np.meshgrid(x, y)

mus = [-0.1, 0, 0.1];

for i, mu in enumerate(mus):
    # Dynamic system
    U = Y + mu*X
    V = -X + mu*Y - Y*X*X
    speed = np.sqrt(U**2 + V**2)

    plt.rc('text', usetex=True)
    plt.rc('font', family='serif')

    fig = plt.figure(num=None, figsize=(30, 30), dpi=80, 
                     facecolor='w', edgecolor='k')
    gs = gridspec.GridSpec(nrows=3, ncols=2, height_ratios=[1, 1, 2])

    ax0 = fig.add_subplot(gs[0, 0])
    ax0.streamplot(X, Y, U, V, density=5, color='k')
    
    # Pontos de equilibrio 
    h_0 = plt.plot(0, 0, 'ro', label='Ponto de equilíbrio')
    
    plt.legend(handles=h_0)
    
    if(mu > 0):
        xmu = np.sqrt((mu**2 + 1)/mu);
        x_mu_m = - xmu;
        x_mu_p = xmu;
        
        y_mu_m = -mu*x_mu_m;
        y_mu_p = -mu*x_mu_p;
        
        plt.plot(x_mu_m, y_mu_m, 'ro')
        plt.plot(x_mu_p, y_mu_p, 'ro')
    
    ax0.set_title(r'Plano de fase - $\dot x = y + \mu x$, $\dot y = -x +\mu y-x^2 y$, $\mu = {}$'.format(mu))
    
    ax0.set_xlabel('$x$');
    ax0.set_ylabel('$y$');

    ax = plt.gca()
    ax.set_xlim(-w, w)
    ax.set_ylim(-h, h)

    plt.gca().set_aspect('equal', adjustable='box')

    plt.savefig('ex2_{}.png'.format(i), bbox_inches="tight", pad_inches=0)
    plt.savefig('ex2_{}.eps'.format(i), bbox_inches="tight", pad_inches=0)
```
